# 🔎 Знакомство с Python

### Типы переменных

Главным объектом в Data Science, конечно, являются данные. С точки зрения python все данные "помещаются" в переменные - это некоторое имя, по которому можно обратиться к определённой области памяти в компьютере, где эти данные хранятся. Имя переменной может быть любым, главное, чтобы оно не начиналось с цифры. Также нужно отметить, что в переменные можно положить объекты разных типов (при этом, в отличие от, например, языка С, С++ указывать тип вручную не нужно, он определится автоматически).

Давайте попробуем создать несколько переменных разных типов:

```python
# создаём переменную-число
number = 10
# переменную-строку
string = 'это строка!' # обязательно пишем в кавычках
# переменную-индикатор (булеву)
flag = True # True или False
# переменную-массив данных
array = [number, string, flag]
print(array)
```

### Функции ввода и вывода

Не всегда удобно заполнять значения переменных в коде (например, когда мы хотим получить что-то от пользователя и затем это обработать, вряд ли мы дадим ему доступ к нашему коду:)). Для этого есть возможность ввода данных с клавиатуры прямо во время работы программы.

Есть также и функции вывода (печати) результатов на экран:

```python
# это функция ввода, она ожидает пока пользователь введёт
# что-нибудь в поле и нажмёт enter
name = input('Введите своё имя: ') # в скобках можно написать подсказку что вводить, но можно и не писать:)
# а это функция вывода
print('Привет,', name)
```

### Арифметические и логические операции

С числами (переменными числовых типов) можно производить обычные математические операции +, -, *, /.

Существуют также логические операции проверки на равенство ==, неравенство != и больше >, меньше <.

```python
# что-нибудь посчитаем
math_expr = (40 * 10 + 2) / 5
print(math_expr)

# сравним результат
print(math_expr == 50)
print(math_expr != 70)
```

### Условные выражения

Удобно составлять условные выражения проверки чего-либо и в случае положительного результата делать какое-нибудь действие. Для этого существуют условные конструкции - давайте посмотрим на пример:

```python
# сначала пишем if, потом логическое условие, потом с отступом тело условия
# если условие истинно, будет выполняться тело
# если ложно, выполнится то, что находится в ветке else
age = int(input('Введите Ваш возраст: ')) # здесь мы явно привели переменную возраста к числовому типу данных
if age > 12:
    print('Добро пожаловать!')
else:
    print('Вход запрещён:(')
# ветку else, кстати, можно не писать
# тогда в случае неистинности условия просто ничего не выполнится
```

```python
# условий можно писать несколько
# проверяться они будут последовательно
ans = input('Хотите пирожок? ')
if ans == 'да' or ans == 'Да' or ans == 'ДА':
    print('Держите:)')
elif (ans == 'нет'):
    print('Ну ладно до скорого:(')
else:
    print('Я Вас не понял')
```

Выше, кстати был показан также пример использования конструкции их нескольких условий в if'е, давайте поподробнее:
- пишем несколько условий, лучше каждое в отдельных скобках
- между ними можно поставить "и" (and), "или" (or)
- можете использовать также отрицание условия - для этого используется not
- PS: скобки можно и не ставить, но тогда нужно учесть, что в этом случае выполнится старндартный приоритет операций: сначала все not, потом and, потом or

### Циклы

И последняя новая тема на сегодня - циклы, Это итеративное повторение одного и того же набора действий, пока не выполнится какое-либо условие; посмотрим на синтаксис:

```python
# сначала пишем ключевое слово while, затем условие, затем тело цикла\
# пока условие истинно, тело цикла будет выполняться
pie_amount = int(input('Введите количество пирожков: '))
while pie_amount < 20:
    print('Это мало, хочу ещё!')
    add = int(input('Введите добавку: '))
    pie_amount = pie_amount + add
print('Спасибо, я наелся:)')
```

PS: while - только одна из разновидностей цикла, в Python есть ещё один вид - цикл for. Но while - более "общий случай": всё, что можно решить через for, можно решить через while; но не всё, что можно решить через while, можно решить через for:)
